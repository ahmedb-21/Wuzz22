<font style="font-family:treasure thin;font-size:30px;color:#000000">- Define(6):</font>
- Convert `salary_minimum` and `salary_maximum` columns' datatype from `float` datatype to `int` datatype.

<font style="font-family:treasure thin;font-size:30px;color:#000000">- Code</font>

```python
#1st., clear all the NaN values:
post_cleaned = post_cleaned.loc[~post_cleaned.salary_minimum.isnull()]
post_cleaned = post_cleaned.loc[~post_cleaned.salary_maximum.isnull()]

#2nd.,Convert the datatypes from 'float' to 'int':
post_cleaned.salary_minimum =post_cleaned.salary_minimum.apply(np.int64)
post_cleaned.salary_maximum = post_cleaned.salary_maximum.apply(np.int64)
```
